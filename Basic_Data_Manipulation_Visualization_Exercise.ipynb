### 30. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
# Step 1: Create a new column 'Male' using np.where to represent whether the gender is Male (1) or not (0)
users['Male'] = np.where(users['gender'] == 'M', 1, 0)

# Step 2: Group the data by "occupation" and sum the 'Male' column to get the total male count and total count
grouped_data = users.groupby('occupation').agg({'Male': ['sum', 'count']})

# Step 3: Calculate the male ratio for each occupation
grouped_data['Male Ratio'] = grouped_data[('Male', 'sum')] / grouped_data[('Male', 'count')]

# Step 4: Sort the results in descending order of the male ratio
sorted_result_df = grouped_data.sort_values(by='Male Ratio', ascending=False)

print(sorted_result_df)
```

### 33.  For each occupation present the percentage of women and men

```python
grouped_data = users.groupby('occupation').agg({'Male': ['sum', 'count']})
grouped_data['female'] = grouped_data['Male']['count'] - grouped_data['Male']['sum']

grouped_data['percentage_men'] = (grouped_data[('Male', 'sum')] / grouped_data[('Male', 'count')] )*100
grouped_data['percentage_women'] = (grouped_data['female'] / grouped_data[('Male', 'count')] )*100

grouped_data
```

### 53. Convert the below Series to pandas datetime : 

DoB = pd.Series(["07Sep59","01Jan55","15Dec47","11Jul42"])

Make sure that the year is 19XX not 20XX

```python
DoB = pd.Series(["07Sep59","01Jan55","15Dec47","11Jul42"])
```

```python

def custom_date(a):
    """
    function to read date from DataFrame & makes the date in 19xx
    """

    date_reader = datetime.datetime.strptime(a,'%d%b%y')
    year_1900 = date_reader.year-100
    my_date = date_reader.replace(year=year_1900)
    return my_date
```

```python
DoB.apply(custom_date)
```
